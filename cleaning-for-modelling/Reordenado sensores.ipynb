```python
import psycopg2 as psy
import numpy as np
import pandas as pd
```

```python
conn = psy.connect(
    host="localhost",
    database="COBI",
    user="postgres",
    password="admin")
```

```python
query = """select time,bbp,cdm from copernicus.bgc  
where (lat between 26.14999999957532 and  26.809166322814903)
and (lon between -113.89444399967512  and -113.26666699967019 )
"""
bgc = pd.read_sql(query, conn)
bgc["time"] = pd.to_datetime(bgc["time"])
bgc = bgc.groupby([pd.Grouper(key='time', freq='D')]).agg(bbp_mean=('bbp', np.mean),
                                                          bbp_max=('bbp', np.max),
                                                          bbp_min=("bbp",np.min),
                                                          bbp_sd=("bbp",np.std),
                                                          cdm_mean=('cdm', np.mean),
                                                          cdm_max=('cdm', np.max),
                                                          cdm_min=("cdm",np.min),
                                                          cdm_sd=("cdm",np.std)
                                                        )


query = """select time,pico,prokar,chl,dino,hapto,green,prochlo,nano,
diato,micro from copernicus.plank  
where (lat between 26.14999999957532 and  26.809166322814903)
and (lon between -113.89444399967512  and -113.26666699967019 )
"""
plank = pd.read_sql(query, conn)
plank["time"] = pd.to_datetime(plank["time"])
plank = plank.groupby([pd.Grouper(key='time', freq='D')]).agg(pico_mean=('pico', np.mean),
                                                          pico_max=('pico', np.max),
                                                          pico_min=("pico",np.min),
                                                          pico_sd=("pico",np.std),
                                                          prokar_mean=('prokar', np.mean),
                                                          prokar_max=('prokar', np.max),
                                                          prokar_min=("prokar",np.min),
                                                          prokar_sd=("prokar",np.std),
                                                          chl_mean=('chl', np.mean),
                                                          chl_max=('chl', np.max),
                                                          chl_min=("chl",np.min),
                                                          chl_sd=("chl",np.std),
                                                          dino_mean=('dino', np.mean),
                                                          dino_max=('dino', np.max),
                                                          dino_min=("dino",np.min),
                                                          dino_sd=("dino",np.std),
                                                          hapto_mean=('hapto', np.mean),
                                                          hapto_max=('hapto', np.max),
                                                          hapto_min=("hapto",np.min),
                                                          hapto_sd=("hapto",np.std),
                                                          green_mean=('green', np.mean),
                                                          green_max=('green', np.max),
                                                          green_min=("green",np.min),
                                                          green_sd=("green",np.std),
                                                          prochlo_mean=('prochlo', np.mean),
                                                          prochlo_max=('prochlo', np.max),
                                                          prochlo_min=("prochlo",np.min),
                                                          prochlo_sd=("prochlo",np.std),
                                                          nano_mean=('nano', np.mean),
                                                          nano_max=('nano', np.max),
                                                          nano_min=("nano",np.min),
                                                          nano_sd=("nano",np.std),
                                                          diato_mean=('diato', np.mean),
                                                          diato_max=('diato', np.max),
                                                          diato_min=("diato",np.min),
                                                          diato_sd=("diato",np.std),
                                                          micro_mean=('micro', np.mean),
                                                          micro_max=('micro', np.max),
                                                          micro_min=("micro",np.min),
                                                          micro_sd=("micro",np.std)
                                                        )
query = """select time, rrs670, rrs412, rrs555, rrs443 from copernicus.reflec  
where (lat between 26.14999999957532 and  26.809166322814903)
and (lon between -113.89444399967512  and -113.26666699967019 )
"""
reflec = pd.read_sql(query, conn)

query = """select time, analysed_sst from copernicus.sst  
where (lat between 26.14999999957532 and  26.809166322814903)
and (lon between -113.89444399967512  and -113.26666699967019 )
"""
sst = pd.read_sql(query, conn)
sst["time"] = pd.to_datetime(sst["time"])
sst = sst.groupby([pd.Grouper(key='time', freq='D')]).agg(sstt_mean=('analysed_sst', np.mean),
                                                          sst_max=('analysed_sst', np.max),
                                                          sst_min=("analysed_sst",np.min),
                                                          sst_sd=("analysed_sst",np.std))

query = """select time,spm, kd490, zsd from copernicus.trans  
where (lat between 26.14999999957532 and  26.809166322814903)
and (lon between -113.89444399967512  and -113.26666699967019 )
"""
trans = pd.read_sql(query, conn)

query = """select time, vo, uo from copernicus.tsc  
where (latitude between 26.14999999957532 and  26.809166322814903)
and (longitude between -113.89444399967512  and -113.26666699967019 )
"""
tsc = pd.read_sql(query, conn)




```

```python
full_obs = pd.concat([bgc,sst])
```

```python
for column in full_obs.columns:
    full_obs[column]= full_obs[column].interpolate("time")
```

```python
full_obs = full_obs.dropna()
```

```python
full_obs.to_csv("sensor_base.csv")
```
